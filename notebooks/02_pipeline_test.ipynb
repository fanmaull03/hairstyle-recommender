```python
import sys
import os

# Arahkan ke root folder proyek agar bisa import dari core/
sys.path.append(os.path.abspath('..'))

# Cek apakah path sudah benar
print(os.listdir('..'))  # harusnya tampil: app, core, data, models, dll
```

```python
from core.recommender import HairstyleRecommender
import cv2
import matplotlib.pyplot as plt

# Path absolut ke root proyek
ROOT = os.path.abspath('..')
rec = HairstyleRecommender(
    model_path=os.path.join(ROOT, 'models', 'shape_predictor_68_face_landmarks.dat'),
    data_path=os.path.join(ROOT, 'data', 'hairstyles.json')
)
result = rec.analyze(os.path.join(ROOT, 'data', 'foto_test.jpeg'))

if result['success']:
    print('Bentuk wajah :', result['label'])
    print('Confidence   :', f"{result['confidence']*100:.1f}%")
    print('Alasan       :', result['reasoning'])
    print()
    print('Fitur geometri:')
    for k, v in result['features'].items():
        print(f'  {k:20s}: {v}')
else:
    print('ERROR:', result['error'])
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(result['annotated_img'], cv2.COLOR_BGR2RGB))
plt.title(f"Bentuk wajah: {result['label']} ({result['confidence']*100:.0f}%)")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
print('Rekomendasi hairstyle:')
for i, h in enumerate(result['hairstyles'], 1):
    print(f"\n{i}. {h['name']}")
    print(f"   {h['description']}")
    print(f"   Cocok untuk rambut: {', '.join(h['hair_type'])}")
    print(f"   Tags: {', '.join(h['tags'])}")
```
